```python
import torch

lb = 19.40  
ub = 26.78  
length = 149

# 生成长度为 length 的均匀分布的随机张量
# torch.manual_seed(1)
random_tensor1 = torch.rand(length)
# torch.manual_seed(2)
random_tensor2 = torch.rand(length)
# torch.manual_seed(3)
random_tensor3 = torch.rand(length)

# 缩放和平移张量到指定的区间 [lb, ub]
depth1 = torch.mul(random_tensor1, (ub - lb)) + lb
depth2 = torch.mul(random_tensor2, (ub - lb)) + lb
depth3 = torch.mul(random_tensor3, (ub - lb)) + lb

print("随机张量1:", depth1)
print("随机张量2:", depth2)
print("随机张量3:", depth3)

# torch.save(scaled_tensor, 'scaled_tensor.pt')
```

```python
torch.save(depth1, './test_depth/depth1.pt')
torch.save(depth2, './test_depth/depth2.pt')
torch.save(depth3, './test_depth/depth3.pt')
```
